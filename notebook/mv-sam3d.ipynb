```python
from plyfile import PlyData
import numpy as np, plotly.graph_objects as go
path="/content/MV-SAM3D/visualization/mv-cup/bottle/mv-cup_bottle_8v_s1a30_s2e30_20260113_210951/result.ply"
v=PlyData.read(path)["vertex"].data
xyz=np.c_[v["x"],v["y"],v["z"]]
rgb=np.clip(np.c_[v["f_dc_0"],v["f_dc_1"],v["f_dc_2"]]*0.28209479177387814+0.5,0,1)
idx=np.random.choice(len(xyz), min(50000,len(xyz)), replace=False); c=(rgb[idx]*255).astype(np.uint8)
go.Figure(go.Scatter3d(x=xyz[idx,0],y=xyz[idx,1],z=xyz[idx,2],mode="markers",
                       marker=dict(size=1,color=[f"rgb({r},{g},{b})" for r,g,b in c]))).update_layout(scene_aspectmode="data").show()
```
